<b>Introducción</b>

Comparación del rendimiento entre Python y Cython para el programa de las órbitas. Se realiza un test con diferentes cargas para ver el tiempo de ejecución con los dos lenguajes de programación.

En principio Cython debe tener un mejor rendimiento que Python, esto es debido a que con Cython podemos hacer uso del control que proporciona el lenguaje C, y con esto mejorar el uso de procesamiento de la computadora y así mejorar el rendimiento.

<b>Procedimiento</b> 

Importación de las librerías

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

Este csv fue obtenido mediante la ejecución del programa. Se realizaron varias pruebas con diferentes cargas para el programa de las órbitas y arrojaron los resultados que se verán a continuación.

```python
narchivo = "orbitTest2.csv"
```

Manejo de tablas mediante pandas, se eliminan las columnas y filas que estaban en nulos, para que sea más sencilla y correcta la graficación de los resultados.

```python
tabla = pd.read_csv(narchivo, sep=';',decimal=',')
```

```python
tabla
```

```python
tabla = tabla.dropna()
```

Como se evidencia en la siguiente tabla, los resultados de cython son muchos mejores en términos de rendimiento que los de Python.

```python
tabla
```

Graficación de los resultados obtenidos mediante matplot. El color rojo representa Python y el azul representa Cython.

```python
x = tabla['n_steps']
y1 = tabla['tiempo_python']
y2 = tabla['tiempo_cython']
plt.plot(x, y1, alpha = 1.5, color = "red")
plt.plot(x, y2, alpha = 1, color = "blue")
plt.title('Comparación rendimiento Python vs Cython')
plt.ylabel('Time')
plt.xlabel('Steps')
plt.grid()
plt.show()
```

<b>Conclusiones</b>



*   El rendimiento de Cython es mucho mejor que el de Python debido a la manera en que se administra los recursos de procesamiento del equipo.
*   En la primera carga el rendimiento es básicamente el mismo, sin embargo, la verdadera diferencia y mejora entre el rendimiento se evidencia cuando se aumentan las cargas, ya que Python ejecuta el programa de la misma manera, mientras que Cython reparte los recursos.
*   **No solo se ve una mejora en el rendimiento cuando se pueden configurar los recursos con los que el computador va a ejecutar un programa, sino también los propios lenguajes de programación se ven afectados por la forma en la que están programados, y algunos tienen mejor rendimiento que otros.**
